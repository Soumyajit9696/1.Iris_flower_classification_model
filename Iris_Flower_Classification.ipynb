## **LOAD THE DATA**

We have uploaded the iris.data csv file in Google Colab. Now we use the **pd.read_csv** function to read the data inside the iris.data.



```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'flower_class']
df = pd.read_csv("iris.data",names=columns)
df.head(150)
```

## **TRAINING THE MODEL**

```python
#Separating input and output columns >>
data = df.values
X = data[:,0:4]
Y = data[:,4]
#print(Y)
```

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
#print(y_train)
```

# *DIFFERENT TYPES OF ALGORITHM*
# **1. Support Vector Machine Alogorithm**
▪ Here we will be using three kinds of different Machine Learning Alogorithms or models  to predict the class of the flower.The models are--


1.   Support Vector Alogorithm
2.   Logistic Regression Algorithm
3.   Decision Tree Classifier Algorithm




              




```python
# Support vector machine algorithm
from sklearn.svm import SVC
svn = SVC()
svn.fit(X_train, y_train)
```

```python
# Predict from the test dataset
prediction1 = svn.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy=(accuracy_score(y_test,prediction1)*100)
print('The accuracy in case of Supprt Vector Machine algorithm is:',accuracy,"%")
```

```python
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

# **2. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression(solver='lbfgs',max_iter=3000)
model_LR.fit(X_train, y_train)
```

```python
prediction2 = model_LR.predict(X_test)
#Calculation of Accuracy
from sklearn.metrics import accuracy_score
print("The accuracy score in case of Logistic Regression is: ",accuracy_score(y_test, prediction2)*100,"%")
```

```python
for i in range(len(prediction2)):
  print(y_test[i],prediction2[i])
```

# **3. Decision Tree Calssifier**

```python
#Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,y_train)
```

```python
prediction3 = model_DTC.predict(X_test)
# calculation of accuracy
from sklearn.metrics import accuracy_score
print("The accuracy in case of Decision tre calssifier is: ",accuracy_score(y_test, prediction3)*100,"%")
```

```python
for i in range(len(prediction3)):
 print(y_test[i],'  ', prediction3[i])
```

### **Detailed Classification Report**

```python
# classificatoion report
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction3))
```
